```python
import os
os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'
from keras import models
import numpy as np
import PIL
import tkinter as tk
from tkinter import filedialog as fd
from tkinter import *
import TKinterModernThemes as TKMT
import sys
file_path = os.path.join(sys._MEIPASS, "model", "eb0.h5")
model = models.load_model(file_path)
iz = (224, 224)
lim = ""
obj = ""
labels = ["Normal", "Mild", "Moderate", "Severe", "Proliferative"]
def Predict():
    img = PIL.Image.open(lim).convert("RGB")
    img = img.resize(iz)
    img = np.asarray(img)
    img = np.expand_dims(img, axis=0)
    pd = model.predict(img, verbose=0)
    res = np.argmax(pd)
    obj.TextObj.config(text=f'Result: {res} - {labels[res]}')
class GUI(TKMT.ThemedTKinterFrame):
    def _init_(self):
        global obj
        super()._init_("AID-R", "sun-valley", "dark")
        obj = self
        self.root.iconbitmap(os.path.join(sys._MEIPASS, "model", "Lerolero.ico"))
        self.UploadFrame = self.addFrame("Menu")
        self.UploadFrame.OptionMenu(["1 - EfficientNetB0", "2 - DenseNet121"], tk.StringVar(value="1 - EfficientNetB0"), lambda x:globals().update(model=models.load_model(os.path.join(sys._MEIPASS, "model", "eb0.h5" if x == "1 - EfficientNetB0" else "dn.h5")), iz = (224, 224) if x == "1 - EfficientNetB0" else (256, 256)))
        self.UploadFrame.Button("Select Image", lambda:(globals().update(lim=fd.askopenfilename(
            initialdir=os.getcwd(),
            title="Select Image",
            filetypes=(("PNG", ".png"), ("JPG", ".jpg"), ("JPEG", "*.jpeg"))
        ))))
        self.UploadFrame.Button("Predict", Predict)
        self.TextObj = self.UploadFrame.Text("Result: ")
        self.run()
GUI()
```
